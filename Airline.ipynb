##### Here we extracted hoir and minute from dep_time column and Arrival_time column

```python
def extract_hour_minute(dff,col):                    
    dff[col+'_hour']=dff[col].dt.hour
    dff[col+'_minute']=dff[col].dt.minute
    dff.drop(col, axis=1,inplace=True)
    return dff.head(2)

```

```python
## here we defined the time slots to see how many early morning, and repsectively
```

```python
#now we will define a function that will convert dep_time of flight into when it left


def flight_dep_time(x):
    
    if(x>4) and (x<=8):
        return 'earnly morning'
    elif(x>8) and (x<=12):
        return 'morning'
    elif(x>12) and (x<=16):
        return 'noon'
    elif(x>16) and (x<=20):
        return 'the evening'
    elif(x>20) and (x<=24):
        return 'night'
    else:
        return 'late night'
        
    
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
reg_rf=RandomForestRegressor()
```

```python
np.linspace(start=1000,stop=1200, num=6)
```

```python
n_estimators=[int(x) for x in np.linspace(start=1000,stop=1200,num=6)]

max_features=["auto", "sqrt"]

max_depth=[int(x) for x in np.linspace(start=5, stop=30, num=4)]

min_samples_split=[5,10,15,100]
```

```python
random_grid={
    'n_estimators':n_estimators,
    'max_features':max_features,
    'max_depth': max_depth,
    'min_samples_split': min_samples_split
}
```

```python
random_grid
```

```python
rf_random=RandomizedSearchCV(reg_rf,param_distributions=random_grid,cv=3,verbose=2, n_jobs=-1)
```

```python
rf_random.fit(X_train, y_train)
```

```python
rf_random.best_params_
```

```python
pred_2=rf_random.predict(X_test)
```

```python
metrics.r2_score(y_test,pred_2)
```
